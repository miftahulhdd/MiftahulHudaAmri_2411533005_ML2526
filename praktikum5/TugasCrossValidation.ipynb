```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score, roc_auc_score)
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')
```

```python
sub_url = 'https://raw.githubusercontent.com/miftahulhdd/MiftahulHudaAmri_2411533005_ML2526/refs/heads/main/praktikum5/gender_submission.csv'
train_url = 'https://raw.githubusercontent.com/miftahulhdd/MiftahulHudaAmri_2411533005_ML2526/refs/heads/main/praktikum5/train.csv'
test_url = 'https://raw.githubusercontent.com/miftahulhdd/MiftahulHudaAmri_2411533005_ML2526/refs/heads/main/praktikum5/test.csv'

train = pd.read_csv(train_url)
test  = pd.read_csv(test_url)
sub   = pd.read_csv(sub_url)

print('Train shape:', train.shape)
print('Test shape :', test.shape)
print('Submission :', sub.shape)
print()
print('Missing values pada train.csv:')
print(train.isnull().sum())
```

```python
FEATURES = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

def preprocess(df):
    df = df.copy()
    df['Sex']      = df['Sex'].map({'male': 1, 'female': 0})
    df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
    X = df[FEATURES]
    imputer = SimpleImputer(strategy='median')
    X = pd.DataFrame(imputer.fit_transform(X), columns=FEATURES)
    return X

# Preprocess train dan test
X_train_all = preprocess(train)
y_train_all  = train['Survived'].values

X_test_all = preprocess(test)
y_test_all  = sub['Survived'].values   # label dari gender_submission.csv

# Scaling — fit HANYA pada train, transform keduanya
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_all)
X_test_scaled  = scaler.transform(X_test_all)    # hanya transform, bukan fit
```

```python
model = LogisticRegression(max_iter=1000, random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

metrics = {
    'Accuracy' : 'accuracy',
    'Precision': 'precision',
    'Recall'   : 'recall',
    'F1 Score' : 'f1',
    'ROC-AUC'  : 'roc_auc'
}

print('=== K-Fold Cross Validation (5-Fold) pada train.csv ===')
for name, scoring in metrics.items():
    scores = cross_val_score(
        model, X_train_scaled, y_train_all, cv=kf, scoring=scoring
    )
    print(f'{name:10s}: {scores.round(4)} | Mean={scores.mean():.4f} | Std={scores.std():.4f}')
```

```python
model.fit(X_train_scaled, y_train_all)

# Prediksi pada test.csv
y_pred = model.predict(X_test_scaled)
y_prob = model.predict_proba(X_test_scaled)[:, 1]

# Bandingkan dengan label dari gender_submission.csv
print('=== Final Holdout Evaluation (test.csv vs gender_submission.csv) ===')
print(f'Accuracy : {accuracy_score(y_test_all,  y_pred):.4f}')
print(f'Precision: {precision_score(y_test_all, y_pred):.4f}')
print(f'Recall   : {recall_score(y_test_all,    y_pred):.4f}')
print(f'F1 Score : {f1_score(y_test_all,        y_pred):.4f}')
print(f'ROC-AUC  : {roc_auc_score(y_test_all,   y_prob):.4f}')
```
